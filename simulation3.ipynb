## Theory:
### What is $\pi$ (Pi)?

Pi ($\pi$) is a mathematical constant that represents the ratio of the circumference of a circle to its diameter.

$$
\pi = \frac{\text{Circumference}}{\text{Diameter}}
$$

Its approximate value is:

$$
\pi \approx 3.14159
$$

### Properties of Pi

- **Symbol:** $\pi$
- **Approximate value:** 3.14159
- It is an **irrational number**, meaning it cannot be expressed as a fraction $\frac{a}{b}$, where $b \neq 0$.
- It is a **non-terminating** and **non-recurring** decimal number.
- Pi is one of the most important mathematical constants used in geometry, engineering, physics, and computer science.

## History of Pi

### 1. Babylonian and Egyptian Civilization

The earliest approximations of $\pi$ were made by the **Babylonians** and **Egyptians**.

The Babylonians used:

$$
\pi \approx 3.125
$$

The Egyptians also developed approximations while solving problems related to **architecture**, **construction**, and **astronomy**.

### 2. Greek Mathematician – Archimedes

The Greek mathematician **Archimedes** introduced the **Polygon Method** to estimate $\pi$.

He inscribed and circumscribed polygons around a circle and calculated their perimeters.

Using this method, he showed that

$$
3.1408 < \pi < 3.1429
$$

This was one of the most accurate estimates of $\pi$ during ancient times.

## Algorithms to Generate Pi

Several mathematical methods can be used to approximate the value of $\pi$.

### 1. Madhava–Leibniz Formula

One of the earliest infinite series for $\pi$ is

$$
\pi = 4\left(1-\frac{1}{3}+\frac{1}{5}-\frac{1}{7}+\frac{1}{9}-\cdots\right)
$$

As more terms are added, the approximation becomes more accurate.

### 2. Wallis Product

John Wallis introduced an infinite product for $\pi$.

$$
\frac{\pi}{2}
=
\frac{2 \times 2}{1 \times 3}
\times
\frac{4 \times 4}{3 \times 5}
\times
\frac{6 \times 6}{5 \times 7}
\times \cdots
$$

Although the Wallis Product converges slowly, it provides another mathematical way of approximating $\pi$.

## Monte Carlo Method for Estimating Pi

The **Monte Carlo Method** is a probabilistic simulation technique that estimates the value of $\pi$ using random numbers.

Instead of calculating $\pi$ directly, random points are generated inside a square that contains an inscribed circle.

### Working Principle

1. Draw a square of side length $2r$.
2. Draw a circle of radius $r$ inside the square.
3. Generate a large number of random points inside the square.
4. Count:
   - Total number of random points generated.
   - Number of points that fall inside the circle.
5. Estimate $\pi$ using the ratio of areas.

Since

$$
\frac{\text{Area of Circle}}{\text{Area of Square}}
=
\frac{\pi r^2}{(2r)^2}
=
\frac{\pi}{4}
$$

Therefore,

$$
\pi
=
4 \times
\frac{\text{Number of Points Inside the Circle}}
{\text{Total Number of Generated Points}}
$$

The accuracy of the estimated value improves as the number of random points increases.

---

## Law of Large Numbers

The Monte Carlo method is based on the **Law of Large Numbers**.

This law states that as the number of random experiments increases, the average of the observed results approaches the expected value.

In Pi simulation, increasing the number of randomly generated points causes the estimated value of $\pi$ to become closer to the actual value of

$$
\pi \approx 3.14159.
$$

## Simulation Method

In this experiment:

1. Generate **10,000 random integers** between **1 and 10**.
2. Compute the cumulative sum after each trial.
3. Calculate the running average using the cumulative sum.
4. Plot the running average against the number of trials.
5. Compare the running average with the theoretical expected value of **5.5**.

---

## Graph Interpretation

The graph displays two lines:

- **Running Average:** The average of all generated random numbers after each trial.
- **Expected Value (5.5):** The theoretical average shown as a horizontal reference line.

At the beginning of the experiment, the running average changes rapidly because of the small sample size. As more trials are performed, the fluctuations become smaller, and the running average stabilizes near **5.5**. This behavior illustrates the **Law of Large Numbers**.

```python
from __future__ import division
from random import random
from math import pi
import matplotlib.pyplot as plt
import sys


def rain_drop(length_of_field=1):
    """
    Simulate a random rain drop within the field.
    """
    return [
        (.5 - random()) * length_of_field,
        (.5 - random()) * length_of_field
    ]


def is_point_in_circle(point, length_of_field=1):
    """
    Check if a point lies within the inscribed circle.
    """
    return (point[0]) ** 2 + (point[1]) ** 2 <= (length_of_field / 2) ** 2


def plot_rain_drops(drops_in_circle, drops_out_of_circle,
                    length_of_field=1, format='png'):
    """
    Plot the raindrops in and out of the circle.
    """
    plt.figure()
    plt.xlim(-length_of_field / 2, length_of_field / 2)
    plt.ylim(-length_of_field / 2, length_of_field / 2)

    plt.scatter(
        [e[0] for e in drops_in_circle],
        [e[1] for e in drops_in_circle],
        color='blue',
        label="Drops in circle"
    )

    plt.scatter(
        [e[0] for e in drops_out_of_circle],
        [e[1] for e in drops_out_of_circle],
        color='black',
        label="Drops out of circle"
    )

    plt.legend(loc="center")
    plt.title(
        f"{len(drops_in_circle) + len(drops_out_of_circle)} drops: "
        f"{len(drops_in_circle)} landed in circle, "
        f"estimating π as "
        f"{4 * len(drops_in_circle) / (len(drops_in_circle) + len(drops_out_of_circle)):.4f}."
    )

    plt.savefig(f"rain_drops.{format}")


def plot_pi_estimates(pi_estimate, total_drops):
    """
    Plot the π estimates as a function of the number of drops.
    """
    plt.figure()

    plt.scatter(range(1, total_drops + 1), pi_estimate, s=1)

    plt.hlines(
        pi,
        0,
        total_drops,
        color='black',
        linestyle='dashed',
        label="Actual π"
    )

    plt.xlim(0, total_drops)
    plt.title("π estimate against number of rain drops")
    plt.xlabel("Number of rain drops")
    plt.ylabel("π")
    plt.legend()

    plt.savefig(f"pi_estimate_{total_drops}_drops.png")


def rain(number_of_drops=1000, length_of_field=1,
         plot=True, dynamic=False):
    """
    Simulate raindrops to estimate π.
    """
    number_of_drops_in_circle = 0
    drops_in_circle = []
    drops_out_of_circle = []
    pi_estimate = []

    for k in range(number_of_drops):

        drop = rain_drop(length_of_field)

        if is_point_in_circle(drop, length_of_field):
            drops_in_circle.append(drop)
            number_of_drops_in_circle += 1
        else:
            drops_out_of_circle.append(drop)

        # Update π estimate
        pi_estimate.append(4 * number_of_drops_in_circle / (k + 1))

        # Optional dynamic plotting
        if dynamic:
            plot_rain_drops(
                drops_in_circle,
                drops_out_of_circle,
                length_of_field,
                format='png'
            )

    # Final plots
    if plot and not dynamic:
        plot_rain_drops(
            drops_in_circle,
            drops_out_of_circle,
            length_of_field
        )

        plot_pi_estimates(pi_estimate, number_of_drops)

    return number_of_drops_in_circle, number_of_drops


if __name__ == "__main__":


    try:

        if len(sys.argv) > 1:
            # Use command-line argument if provided
            number_of_drops = int(sys.argv[1])
        else:
            # Ask user for input
            number_of_drops = int(input("Enter the number of drops: "))

        # Simulate raindrops
        result = rain(number_of_drops, plot=True, dynamic=False)

        print("----------------------")
        print(f"{number_of_drops} drops")
        print(f"π estimated as: {4 * result[0] / result[1]:.6f}")
        print("----------------------")

    except ValueError:
        print("Invalid input! Please provide an integer for the number of drops.")
```

```python
"""
Law of Large Numbers Demo
--------------------------
We roll a random number between 1 and 10 (like a 10-sided die), 10,000 times.
After each roll we compute the RUNNING AVERAGE of all rolls so far.

Theory says: as the number of trials -> infinity, the running average
should converge to the true expected value:

    E[X] = (1 + 2 + ... + 10) / 10 = 5.5

The graph shows the running average settling down near 5.5 as the
number of trials grows.
"""

import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# Step 1: Generate 10,000 random integers between 1 and 10
# ---------------------------------------------------------
np.random.seed(42)          # remove this line if you want different results each run
N = 10000
rolls = np.random.randint(1, 11, size=N)   # random integers in [1, 10]

# ---------------------------------------------------------
# Step 2: Compute the running (cumulative) average at each step
# ---------------------------------------------------------
cumulative_sum = np.cumsum(rolls)
trial_numbers = np.arange(1, N + 1)
running_average = cumulative_sum / trial_numbers

# ---------------------------------------------------------
# Step 3: Print some summary numbers
# ---------------------------------------------------------
true_mean = (1 + 10) / 2   # = 5.5 for a uniform 1..10 distribution
final_average = running_average[-1]

print(f"Theoretical expected value : {true_mean}")
print(f"Average after 10 trials    : {running_average[9]:.4f}")
print(f"Average after 100 trials   : {running_average[99]:.4f}")
print(f"Average after 1,000 trials : {running_average[999]:.4f}")
print(f"Average after 10,000 trials: {final_average:.4f}")

# ---------------------------------------------------------
# Step 4: Plot the running average vs number of trials
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.plot(trial_numbers, running_average, color="steelblue", linewidth=1.2,
         label="Running average of rolls")
plt.axhline(y=true_mean, color="red", linestyle="--", linewidth=1.5,
            label=f"True mean = {true_mean}")

plt.xscale("log")   # log scale makes early fluctuations & convergence easier to see
plt.xlabel("Number of trials (log scale)")
plt.ylabel("Running average")
plt.title("Law of Large Numbers: Random numbers 1–10, up to 10,000 trials")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.savefig("law_of_large_numbers.png", dpi=150)
plt.show()
print(f"Actual π: {pi:.6f}")
print("----------------------")
```
